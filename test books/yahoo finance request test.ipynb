```python
import requests

# Alpha Vantage API parameters
symbol = 'AAPL'
api_key = 'V2LEFQPR1DX9XAFH'  # Replace with your Alpha Vantage API key
interval = '5min'
start_date = '2018-01-01'
end_date = '2019-12-31'

# Build the API URL
url = f'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol={symbol}&interval={interval}&outputsize=full&apikey={api_key}'

# Send the request and retrieve the data
response = requests.get(url)
data = response.json()

# Access the 5-minute interval data for the specified week
time_series = data['Time Series (5min)']
for date, values in time_series.items():
    if start_date <= date <= end_date:
        print(f'Date: {date}')
        print('Open:', values['1. open'])
        print('High:', values['2. high'])
        print('Low:', values['3. low'])
        print('Close:', values['4. close'])
        print('Volume:', values['5. volume'])
        print('-------------------')
```

```python
import pandas as pd

# Assuming 'time_series' contains the data obtained from Alpha Vantage

# Convert the 'time_series' dictionary to a DataFrame
df = pd.DataFrame.from_dict(time_series, orient='index')

# Convert the column names to a more descriptive format
df.columns = ['Open', 'High', 'Low', 'Close', 'Volume']

# Convert the index (date) to a datetime object
df.index = pd.to_datetime(df.index)

# Sort the DataFrame by the index (date) in ascending order
df.sort_index(ascending=True, inplace=True)

# Print the resulting DataFrame
print(df)
df.to_csv("C:\\Users\\Fabio\\OneDrive\\Documents\\Studies\\Final Project\\Social-Media-and-News-Article-Sentiment-Analysis-for-Stock-Market-Autotrading\\test.csv")
```
